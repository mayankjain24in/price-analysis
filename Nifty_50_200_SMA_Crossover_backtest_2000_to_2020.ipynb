```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
%matplotlib inline
```

```python
df = pd.read_excel(r'C:\Users\ABC\Downloads\nifty - 2000 to 2020.xlsx', usecols=['Date','Close'])
```

```python
df.head(2)
```

```python
sma1 = 50
sma2 = 200

df['50SMA'] = np.round(df['Close'].rolling(window=sma1).mean(),2)
df['200SMA'] = np.round(df['Close'].rolling(window=sma2).mean(),2)
```

```python
df.tail()
```

```python
fig = px.line(df, x = 'Date', y = 'Close', title = 'NIfty 50 & 200 SMA Crossover', template = 'simple_white')
fig.add_scatter(x = df['Date'], y = df['50SMA'], name = '50SMA', mode = 'lines' )
fig.add_scatter(x = df['Date'], y = df['200SMA'], name = '200SMA', mode  ='lines')
fig.show()
```

```python
df2 = df.copy()
```

```python
df2['Signal'] = np.where((df2['50SMA'] > df2['200SMA']) & (df2['50SMA'].shift(1) < df2['200SMA'].shift(1)), 1, 0)
df2['Signal'] = np.where((df2['50SMA'] < df2['200SMA']) & (df2['50SMA'].shift(1) > df['200SMA'].shift(1)), -1, df2['Signal'])
```

```python
df2['Signal'].value_counts()
```

```python
df2['Signal'] = df2['Signal'].replace(to_replace=0, method = 'ffill')
```

```python
df2['Hold_Strategy_Return'] = df2['Close'] - df2['Close'].shift(1)
```

```python
df2['Strategy_Return'] = df2['Signal'].shift(1) * df2['Hold_Strategy_Return']
```
